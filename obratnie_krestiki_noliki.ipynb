Можно инициализировать игровое поле любого размера, задав значение N. Можно выбрать последовательность крестиков и ноликов любой длины не больше чем N, задав параметр numx и num0 .

```python
# Поиск диагональных вхождений
def search_diag_enti(enti):
    global_stop = 0
    if enti == numx:
        s = 'User'
    else:
        s = 'Deepmind'

    # df2 = df.copy(deep=True)
    # df3 = df.copy(deep=True)

    # проверяем главную диагональ и все справа от нее 
    for j in range(N):
        arr = ''
        for i in range(N-j): # длина диагонали уменьшается с каждым шагом, чтобы не было выхода за предел массива    
            arr += df[i+j][i]        
        if enti in arr: # и проверяем условие конца игры
            print(f'{s} проиграл. Кoнец игры')
            global_stop = 1
            break

    # проверяем все диагонали слева от главной     
    for j in range(N-1):
        arr = ''
        for i in range(N-1-j):        
            arr += df[i][i+j+1]
        # и проверяем условие конца игры
        if enti in arr:
            print(f'{s} проиграл. Кoнец игры')
            global_stop = 1
            break

    return global_stop        

# search_diag_enti()
```

```python
# Проверка по обратным диагоналям
def search_reverse_diag_enti(enti):
    global_stop = 0
    if enti == numx:
        s = 'User'
    else:
        s = 'Deepmind'

    # df2 = df.copy(deep=True)
    # df3 = df.copy(deep=True)

    # проверяем главную диагональ и все справа от нее 
    for j in range(N): # номер диагонали
        arr = ''
        for i in range((N-j-1),-1,-1): # длина диагонали уменьшается с каждым шагом, чтобы не было выхода за предел массива    
            arr += df[i+j][N-i-1]     
        if enti in arr: # и проверяем условие конца игры
            print(f'{s} проиграл. Кoнец игры')
            global_stop = 1
            break

    # проверяем все диагонали слева от главной     
    for j in range(N-1):
        arr = ''
        for i in range(N-j-2, -1, -1):        
            arr += df[i][N-i-2-j] # !!!
        # и проверяем условие конца игры
        if enti in arr:
            print(f'{s} проиграл. Кoнец игры')
            global_stop = 1
            break

    return global_stop        

```

```python
# Поиск вхождений по столбцам и строкам

def search_axis_enti(enti):
    global_stop = 0
    if enti == numx:
        s = 'User'
    else:
        s = 'Deepmind'
    
    arr1 = df.sum(axis=1) # ищем вхождения по строкам
    arr2 = df.sum(axis=0) # ищем вхождения по столбцам
    
    for i in df:
        if enti in arr1[i] or enti in arr2[i]:
            print(f'{s} проиграл. Кoнец игры')
            global_stop = 1
            # return break # прерывание
        # else:
        #     continue # продолжаем игру
    return global_stop
```

```python

# Инициализируем поле
N=10
numx = 'xxxxx'
num0 = '00000'
df = pd.DataFrame('.', index=range(N), columns=range(N))
print(df)

# инициализируем список всех позиций
positions = []
for i in range(N):
    for j in range(N):
        positions.append((i,j))

# Главный цикл программы
count = 0
while count < (N*N/2):
    count += 1
    
    user_turn()
    # проверка условия поражения
    if search_axis_enti(numx) == 1:
        break
    if search_diag_enti(numx) == 1:
        break
    if search_reverse_diag_enti(numx) == 1:
        break

    deepmind_turn()
    # проверка условия поражения
    if search_axis_enti(num0) == 1:
        break
    if search_diag_enti(num0) == 1:
        break
    if search_reverse_diag_enti(num0) == 1:
        break

    if count == (N*N/2 - 1):
        print('Ничья')
        break
```
